## Matplotlib

3) Create a 1D numpy array with 1000 elements that has random integer values between 0 and 20.

```python
aa = np.random.randint(0, 20, size=(1000,))
```

4) Plot a histogram of the array created in previous exercise

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(aa);
```

5) Create another array similar to the one in exercise three, this time values should be between 10 and 30.

```python
bb = np.random.randint(10, 30, size=(1000,))
```

6) By the means of visualization check if there is any correlation between the arrays created in exercises (3) and (5).

```python
plt.scatter(aa, bb);
```

7) Create a figure with two subplots (side-by-side) and plot the two arrays versus each other (scatter plot) on the left subplot and the histogram of both of them on the right subplot.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3), dpi=300)

ax1.scatter(aa, bb)

ax2.hist(aa, label="AA")
ax2.hist(bb, alpha=.6, label="BB");
ax2.legend()

fig.tight_layout()
fig.savefig("day4_fig1.png", bbox_inches="tight")
```
